```python
class TrafficDevice:
    def activate(self):
        print("Traffic device activated.")
class TrafficLight(TrafficDevice):
    def activate(self):
        print("Traffic Light: Switching to GREEN.")

class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera: Capturing speeding vehicles.")

class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian Signal: WALK signal ON.")

# New class
class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren: Warning! Emergency vehicle approaching.")
devices = [
    TrafficLight(),
    SpeedCamera(),
    PedestrianSignal(),
    EmergencySiren()
]
for device in devices:
    device.activate()
```
